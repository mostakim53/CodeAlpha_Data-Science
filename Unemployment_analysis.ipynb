```python
#Clean data
# Drop rows with missing values
# Convert Date column to datetime
df[" Date"] = pd.to_datetime(df[" Date"], dayfirst=True, errors="coerce")
df = df.dropna()
df

```

```python
#Region_Countplot
plt.figure(figsize=(10,15))
sns.countplot(y='Region',data=df)
plt.show()
```

```python
#Average,High,lowest unemployment rate
avg_unemployment_rate=df.groupby('Region')[' Estimated Unemployment Rate (%)'].mean()
state_with_highest_unemployment=avg_unemployment_rate.idxmax()
high_unemployment_rate=avg_unemployment_rate.max()
state_with_lowest_unemployment=avg_unemployment_rate.idxmin()
low_unemploy_rate=avg_unemployment_rate.min()
print(f"state  with  high employment : {state_with_highest_unemployment}")
print(f"high employment rate : {high_unemployment_rate}")
print(f"state  with  low employment : {state_with_lowest_unemployment}")
print(f"low employment rate : {low_unemploy_rate}")
```

```python
# Flag data from April 2020 onwards as COVID-19 period
df["COVID_19_Flag"] = "No"
df.loc[df[" Date"] >= "2020-04-01", "COVID_19_Flag"] = "Yes"
# Average unemployment rate before and after COVID-19
df.groupby("COVID_19_Flag")[" Estimated Unemployment Rate (%)"].mean()


```

```python
#Show Unemployment Rate Before and After COVID-19
plt.figure(figsize=(8, 6))
sns.boxplot(x='COVID_19_Flag', y=' Estimated Unemployment Rate (%)', data=df, palette=['#1f77b4', '#ff7f0e'])
plt.title('Unemployment Rate Before and After COVID-19', fontsize=16, fontweight='bold')
plt.xlabel('Period', fontsize=14, fontweight='bold')
plt.ylabel('Unemployment Rate (%)', fontsize=14, fontweight='bold')
plt.xticks([0, 1], ['Pre-COVID', 'Post-COVID'], fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
#Show Average Unemployment Rate by Region
regions=avg_unemployment_rate.index
avg_rates=avg_unemployment_rate.values
plt.bar(regions,avg_rates, color='blue')
plt.xlabel('Region')
plt.ylabel('Average Unemployment Rate')
plt.title('Average Unemployment Rate by Region')
plt.xticks(rotation=90)  
plt.tight_layout()  
plt.show()
```

```python
# Group by Date and take average unemployment rate
national = df.groupby(" Date")[" Estimated Unemployment Rate (%)"].mean()
#Show National Unemployment Rate Over Time
plt.figure(figsize=(12,5))
plt.plot(national.index, national.values, marker="o")
plt.axvline(pd.to_datetime("2020-03-01"), color="red", linestyle="--", label="COVID Start")
plt.title("National Unemployment Rate Over Time")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.legend()
plt.show()
```

```python
# Compare unemployment in Rural vs Urban
area_trend = df.groupby([" Date","Area"])[" Estimated Unemployment Rate (%)"].mean().reset_index()

plt.figure(figsize=(12,5))
sns.lineplot(data=area_trend, x=" Date", y=" Estimated Unemployment Rate (%)", hue="Area")
plt.axvline(pd.to_datetime("2020-03-01"), color="red", linestyle="--", label="COVID Start")
plt.title("Rural vs Urban Unemployment")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
# Before and after March 2020
df["Period"] = df[" Date"].apply(lambda x: "Pre-COVID" if x < pd.to_datetime("2020-03-01") else "COVID/Post")

state_change = df.groupby(["Region","Period"])[" Estimated Unemployment Rate (%)"].mean().reset_index()
state_change = state_change.pivot(index="Region", columns="Period", values=" Estimated Unemployment Rate (%)")
state_change["Change"] = state_change["COVID/Post"] - state_change["Pre-COVID"]

# Show top 10 states with highest increase
state_change.sort_values("Change", ascending=False).head(10)
```

```python
# Create a new column to mark before and after March 2020
df["Period"] = ["Pre-COVID" if d < pd.to_datetime("2020-03-01") else "COVID/Post" for d in df[" Date"]]

# Find average unemployment by region and period
avg_rates = df.groupby(["Region", "Period"])[" Estimated Unemployment Rate (%)"].mean().reset_index()

# Make table with Pre-COVID and COVID/Post side by side
table = avg_rates.pivot(index="Region", columns="Period", values=" Estimated Unemployment Rate (%)")

# Calculate change
table["Change"] = table["COVID/Post"] - table["Pre-COVID"]

#Show top 10 regions with largest increase
top10 = table.sort_values("Change", ascending=False).head(10)
print(top10)
```

```python
# Unemployment Trend Over Time

# Make a line plot of unemployment rate by date
plt.figure(figsize=(14,6))

plt.plot(df[" Date"], df[" Estimated Unemployment Rate (%)"], 
         color="blue", linewidth=2, label="Unemployment Rate")

# Highlight Covid period (March 2020 - Dec 2021) with a shaded area
plt.axvspan(pd.to_datetime("2020-03-01"), 
            pd.to_datetime("2021-12-31"), 
            color="red", alpha=0.2, label="Covid Period")

plt.title("Unemployment Rate Over Time", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.legend()
plt.show()
```

```python
# Yearly Average Unemployment

# Create a new column 'Year' from the Date
df['Year'] = df[' Date'].dt.year

# Calculate average unemployment for each year
yearly_avg = df.groupby('Year')[" Estimated Unemployment Rate (%)"].mean()

# Plot the yearly averages
plt.figure(figsize=(10,6))
plt.bar(yearly_avg.index, yearly_avg.values, color="skyblue")
plt.title("Average Unemployment Rate by Year", fontsize=16)
plt.xlabel("Year")
plt.ylabel("Average Unemployment Rate (%)")
plt.show()
```

```python
# Monthly Seasonal Patterns

# Create a new column 'Month' from the Date
df['Month'] = df[' Date'].dt.month

# Calculate average unemployment for each month
monthly_avg = df.groupby('Month')[" Estimated Unemployment Rate (%)"].mean()

# Plot the monthly averages
plt.figure(figsize=(10,6))
plt.plot(monthly_avg.index, monthly_avg.values, marker="o", color="green")
plt.title("Average Unemployment Rate by Month", fontsize=16)
plt.xlabel("Month")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(range(1,13))  # show months 112
plt.show()
```

```python
# Regional Comparison During Covid 

# Select only Covid years (2020 and 2021)
covid_df = df[(df['Year'] >= 2020) & (df['Year'] <= 2021)]

# Calculate average unemployment by region
region_avg = covid_df.groupby("Region")[" Estimated Unemployment Rate (%)"].mean()

# Sort from highest to lowest
region_avg = region_avg.sort_values(ascending=False)

# Plot the results
plt.figure(figsize=(8,10))
plt.barh(region_avg.index, region_avg.values, color="orange")
plt.title("Average Unemployment Rate by Region (Covid Years 2020-2021)", fontsize=16)
plt.xlabel("Unemployment Rate (%)")
plt.ylabel("Region")
plt.show()
```
